#Vizualização de dados usadno o Matplotlib

##Importando os dados

```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
df = pd.read_csv('monitoramento_tempo.csv')
df.head()
```

```python
df.info()
```

```python
df['data'] = pd.to_datetime(df['data'])
```

```python
df.info()
```

#1º Vizualização

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'], color='g') #lw=4 ls='dotted' marker='o'
eixo.set_title('Temperatura no momento', fontsize=25)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
```

##Vizualização por período

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'], color='g')
eixo.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
eixo.set_title('Temperatura no momento', fontsize=25, pad=20)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.grid(True)
```

```python
fig = plt.figure(figsize=(15,8))

eixo = fig.add_axes([0,0,1,1])
eixo2 = fig.add_axes([0.7, 0.65, 0.3, 0.3])

eixo.plot(df['data'], df['temperatura'], color='g')
eixo.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
eixo.set_title('Temperatura em 2014', fontsize=25, pad=20)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.grid(True)

eixo2.plot(df['data'], df['temperatura'], color='b')
eixo2.set_title('Temperatura no momento', fontsize=15, pad=20)
eixo2.set_ylabel('Temperatura', fontsize=10)
eixo2.set_xlabel('Data', fontsize=10)
eixo2.legend(['Temperatura'], loc='best', fontsize=10)
eixo2.grid(True)
```

##Grafico para um mês

```python
fig = plt.figure(figsize=(15,8))

eixo = fig.add_axes([0,0,1,1])
eixo2 = fig.add_axes([0.7, 0.65, 0.3, 0.3])

eixo.plot(df['data'], df['temperatura'], color='g')
eixo.set_xlim(datetime.datetime(2014,5,1),datetime.datetime(2014,6,1))
eixo.set_ylim(270, 320)
eixo.set_title('Temperatura em maio de 2014', fontsize=25, pad=20)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.grid(True)

azul_esq = df['data'] < datetime.datetime(2014,5,1)
azul_dir = df['data'] > datetime.datetime(2014,6,1)

eixo2.plot(df['data'], df['temperatura'], color='g')
eixo2.plot(df[azul_esq]['data'], df[azul_esq]['temperatura'], color='b')
eixo2.plot(df[azul_dir]['data'], df[azul_dir]['temperatura'], color='b')
eixo2.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
eixo2.set_title('Temperatura em 2014', fontsize=25, pad=20)
eixo2.set_ylabel('Temperatura', fontsize=10)
eixo2.set_xlabel('Data', fontsize=10)
eixo2.legend(['Temperatura'], loc='best', fontsize=10)
eixo2.grid(True)
```

##Vizualização com linhas

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(df['data'], df['temperatura'], color='g')
eixo.set_title('Temperatura no momento', fontsize =25, pad=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.legend(['Temperatura'], fontsize=15, loc='lower right')
eixo.grid(True)

eixo.axhline(max(df['temperatura']), color='r', ls='--')
eixo.axhline(min(df['temperatura']), color='r', ls='--')

x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df['temperatura'])
x11 = df['data'][df['temperatura'].idxmax()-7000]
y11 = max(df['temperatura']) - 5
eixo.annotate('Máximo', xy=(x1,y1), fontsize=20, color='r', xytext=(x11,y11),
              arrowprops=dict(facecolor='k'))

x2 = df['data'][df['temperatura'].idxmin()]
y2 = min(df['temperatura'])
x22 = df['data'][df['temperatura'].idxmin()-7000]
y22 = min(df['temperatura']) + 3
eixo.annotate('Mínimo', xy=(x2,y2), fontsize=20, color='r', xytext=(x22,y22),
              arrowprops=dict(facecolor='k'))
```

##Vizualização por dia da semana

```python
temperatura_semana = df.groupby('dia_da_semana')['temperatura'].mean()
dias = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
temperatura_semana = temperatura_semana[dias]
temperatura_semana
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

#indice = range(len(temperatura_semana))
indice = dias
cores = ['black', 'r', 'g', 'b', 'yellow', 'orange', 'magenta']

eixo.bar(indice, temperatura_semana, color= cores, edgecolor='black')
eixo.set_title('Tempertaura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=10)
eixo.set_ylabel('Temperatura', fontsize=10)
#eixo.set_xticks(indice)
#eixo.set_xticklabels(dias)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

explodir = [ 0.1, 0, 0, 0, 0, 0.1, 0.1]

eixo.pie(temperatura_semana, labels=temperatura_semana.index, autopct='%.1f%%',
         explode=explodir, shadow=True)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
```

#Dados de iris

```python
df = pd.read_csv('iris.csv')
df.head()
```

```python
fig = plt.figure(figsize=(12,6))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.scatter(df['comprimento_sépala'], df['largura_sépala'])
eixo.set_title('Gráfico de dispersão', fontsize=20, pad=15)
eixo.set_xlabel('Comprimento', fontsize=16)
eixo.set_ylabel('Largura', fontsize=16)
eixo.tick_params(labelsize=15)
eixo
```

```python
fig = plt.figure(figsize=(12,6))
eixo = fig.add_axes([0, 0, 1, 1])

cores = {'Iris-setosa':'r', 'Iris-versicolor':'b', 'Iris-virginica':'g'}
marcador = {'Iris-setosa':'x', 'Iris-versicolor':'o', 'Iris-virginica':'v'}

for especie in df['espécie'].unique():
    tmp = df[df['espécie'] == especie]
    eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'],
                 color=cores[especie], marker=marcador[especie], s=100)

eixo.set_title('Gráfico de dispersão', fontsize=20, pad=15)
eixo.set_xlabel('Comprimento', fontsize=16)
eixo.set_ylabel('Largura', fontsize=16)
eixo.tick_params(labelsize=15)
eixo.legend(cores, fontsize=20)
eixo
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0, 0, 1, 1])

cores = ['red', 'blue', 'orange', 'green']

caixas = eixo.boxplot(df.drop('espécie', axis=1).values, patch_artist=True)
eixo.set_title('Boxplot da largura da pétala', fontsize=20)
eixo.set_xlabel('Largura', fontsize=16)

for caixa, cor in zip(caixas['boxes'], cores):
    caixa.set(color=cor)

for outlier in caixas['fliers']:
    outlier.set(marker='x', markersize=8)
    
eixo
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.hist(df['comprimento_pétala'], bins=20, orientation='horizontal', density=True)
eixo.set_title('Histograma', fontsize=20, pad=15)
eixo.set_xlabel('Comprimento', fontsize=16)
eixo.set_ylabel('Descidade', fontsize=16)
eixo.grid(True)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0, 0, 1, 1])

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins=20)
eixo.set_title('Histograma', fontsize=20, pad=10)
eixo.set_xlabel('Comprimento', fontsize=16)
eixo.set_ylabel('Frequência', fontsize=16)
eixo.grid(True)

eixo.annotate('$mu={0:.2f}$\n$sigma={1:.2f}$'.format(mu, sigma),
              xy=(4.5,20), fontsize=20)

eixo.axvline(mu, color='k', linestyle='--')
eixo.annotate('Média', xy=(mu-0.9, 28), fontsize=15)
eixo.axvline(df['comprimento_pétala'].median(), color='k', linestyle='--')
eixo.annotate('Mediana', xy=(df['comprimento_pétala'].median(), 32), fontsize=15)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0, 0, 1, 1])

df_iv = df[df['espécie']=='Iris-versicolor']

mu, sigma = df_iv['comprimento_pétala'].mean(), df_iv['comprimento_pétala'].std()

eixo.hist(df_iv['comprimento_pétala'], bins=20)
eixo.set_title('Iris-versicolor', fontsize=20, pad=10)
eixo.set_xlabel('Comprimento', fontsize=16)
eixo.set_ylabel('Frequência', fontsize=16)
eixo.grid(True)

eixo.annotate('$mu={0:.2f}$\n$sigma={1:.2f}$'.format(mu, sigma),
              xy=(4.6,6), fontsize=20)

eixo.axvline(mu, color='k', linestyle='--')
eixo.annotate('Média', xy=(mu-0.5, 5.5), fontsize=15)
eixo.axvline(df_iv['comprimento_pétala'].median(), color='g', linestyle='--')
eixo.annotate('Mediana', xy=(df_iv['comprimento_pétala'].median()-0.7, 6.5), fontsize=15, color='g')

fig.savefig('Histograma_iv.png', bbox_inches='tight')
```
